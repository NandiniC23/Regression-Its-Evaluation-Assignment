#**Question 1: What is Simple Linear Regression?**
-  Simple Linear Regression
Simple Linear Regression (SLR) is a statistical method used to model the relationship between a single independent variable (predictor) and a continuous dependent variable (target). The goal is to create a linear equation that best predicts the value of the target variable based on the predictor variable.

Equation
The simple linear regression equation is:

Y = β0 + β1X + ε

Where:

- Y: Dependent variable (target)
- X: Independent variable (predictor)
- β0: Intercept or constant term
- β1: Slope coefficient
- ε: Error term

Applications

Simple linear regression is widely used in various fields, such as:

- Predicting continuous outcomes: SLR can be used to predict continuous outcomes like stock prices, temperatures, or energy consumption.
- Analyzing relationships: SLR helps analyze the relationship between two variables, such as the relationship between hours studied and exam scores.

#**Question 2: What are the key assumptions of Simple Linear Regression?**
-  Key Assumptions of Simple Linear Regression
For Simple Linear Regression (SLR) to be applicable and produce reliable results, certain assumptions must be met:

1. Linearity
- Assumption: The relationship between the independent variable (X) and the dependent variable (Y) should be linear.
- Check: Use scatter plots to visualize the relationship between X and Y.

2. Independence
- Assumption: Each observation should be independent of the others.
- Check: Ensure that the data collection process does not introduce dependencies between observations.

3. Homoscedasticity
- Assumption: The variance of the error term should be constant across all levels of X.
- Check: Use residual plots to check for constant variance.

4. Normality
- Assumption: The error term should be normally distributed.
- Check: Use Q-Q plots or statistical tests (e.g., Shapiro-Wilk test) to check for normality.

5. No Multicollinearity (Implicitly Met)
- Assumption: Since there's only one predictor in SLR, multicollinearity is not a concern.

#**Question 3: What is heteroscedasticity, and why is it important to address in regression models?**
-  Heteroscedasticity -

Heteroscedasticity refers to the condition in which the variance of the error term in a regression model is not constant across all levels of the independent variable(s). This means that the spread of the residuals (errors) varies as the values of the independent variable(s) change.

Why is Heteroscedasticity Important to Address?
Heteroscedasticity can lead to:

- Inefficient Estimates: Heteroscedasticity can result in inefficient estimates of regression coefficients, making it difficult to accurately predict the dependent variable.
- Biased Standard Errors: Heteroscedasticity can lead to biased standard errors, which can affect the validity of hypothesis tests and confidence intervals.
- Incorrect Inference: Ignoring heteroscedasticity can lead to incorrect conclusions about the relationships between variables.

Consequences of Ignoring Heteroscedasticity Ignoring heteroscedasticity can result in:

- Overestimation or Underestimation: Heteroscedasticity can lead to overestimation or underestimation of the significance of regression coefficients.
- Poor Model Performance: Heteroscedasticity can negatively impact the performance of the regression model, leading to poor predictions and unreliable results.

#**Question 4: What is Multiple Linear Regression?**
-  Multiple Linear Regression
Multiple Linear Regression (MLR) is a statistical method used to model the relationship between multiple independent variables (predictors) and a continuous dependent variable (target). The goal is to create a linear equation that best predicts the value of the target variable based on the predictor variables.

Equation
The multiple linear regression equation is:

Y = β0 + β1X1 + β2X2 + … + βnXn + ε

Where:

- Y: Dependent variable (target)
- X1, X2, …, Xn: Independent variables (predictors)
- β0: Intercept or constant term
- β1, β2, …, βn: Slope coefficients for each predictor
- ε: Error term

Applications

Multiple linear regression is widely used in various fields, such as:

- Predicting continuous outcomes: MLR can be used to predict continuous outcomes like house prices, stock prices, or energy consumption.
- Analyzing relationships: MLR helps analyze the relationships between multiple predictors and a continuous outcome variable.

#**Question 5: What is polynomial regression, and how does it differ from linear regression?**
-  Polynomial Regression
Polynomial regression is a type of regression analysis where the relationship between the independent variable (X) and the dependent variable (Y) is modeled as a polynomial equation. This allows for more complex, non-linear relationships between the variables.

Key Differences -

| Aspect           | Linear Regression                         | Polynomial Regression                                         |
| ---------------- | ----------------------------------------- | ------------------------------------------------------------- |
| **Equation**     | Straight line ($y = \beta_0 + \beta_1 x$) | Curved line ($y = \beta_0 + \beta_1 x + \beta_2 x^2 + \dots$) |
| **Shape of Fit** | Straight line                             | Can be parabolic, cubic, wavy, etc.                           |
| **Use Case**     | When data shows a linear trend            | When data shows a curved/non-linear trend                     |
| **Complexity**   | Simple, easy to interpret                 | More complex, risk of overfitting if degree is too high       |
| **Computation**  | Uses original features                    | Uses higher-order transformations of features                 |





#Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
#● X = [1, 2, 3, 4, 5]
#● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
#Plot the regression line over the data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Fit Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Plot data points and regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()

# Output slope and intercept
print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)
```

#Question 7: Fit a Multiple Linear Regression model on this sample data:
#● Area = [1200, 1500, 1800, 2000]
#● Rooms = [2, 3, 3, 4]
#● Price = [250000, 300000, 320000, 370000]
#Check for multicollinearity using VIF and report the results.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample data
Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

# Create DataFrame
df = pd.DataFrame({"Area": Area, "Rooms": Rooms, "Price": Price})

# Features and target
X = df[["Area", "Rooms"]]
y = df["Price"]

# Fit Multiple Linear Regression
model = LinearRegression()
model.fit(X, y)

# Regression coefficients
print("Intercept (β0):", model.intercept_)
print("Coefficient for Area (β1):", model.coef_[0])
print("Coefficient for Rooms (β2):", model.coef_[1])

# Add constant for statsmodels VIF calculation
X_const = sm.add_constant(X)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i)
                   for i in range(X_const.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)
```

#Question 8: Implement polynomial regression on the following data:
#● X = [1, 2, 3, 4, 5]
#● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
#Fit a 2nd-degree polynomial and plot the resulting curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform features into polynomial (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values for smooth curve
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
Y_pred = model.predict(poly.transform(X_range))

# Plot data points and polynomial regression curve
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_range, Y_pred, color='red', linewidth=2, label='Polynomial Regression (deg=2)')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd-Degree Polynomial Regression")
plt.legend()
plt.show()

# Output coefficients
print("Intercept (β0):", model.intercept_)
print("Coefficients:", model.coef_)
```

#Question 9: Create a residuals plot for a regression model trained on this data:
#● X = [10, 20, 30, 40, 50]
#● Y = [15, 35, 40, 50, 65]
#Assess heteroscedasticity by examining the spread of residuals.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predictions
y_pred = model.predict(X)

# Residuals
residuals = Y - y_pred

# Plot residuals vs predicted values
plt.scatter(y_pred, residuals, color='blue', marker='o')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()

# Print residuals to check spread
print("Residuals:", residuals)
```

#**Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.**
-  We are predicting house prices using regression, but you’ve detected two issues:

Heteroscedasticity → error variance is not constant (residuals spread out as price increases).

Multicollinearity → predictors (e.g., area & number of rooms, or location dummies) are highly correlated.

Step-by-Step Plan
1. Handling Heteroscedasticity

Heteroscedasticity violates regression assumptions → unreliable standard errors, misleading p-values.

Solutions:

Transformation of dependent variable

Apply log-transformation on Price (e.g., log(price) ~ features).

Often house prices follow a log-normal distribution → this stabilizes variance.

Transformation of features

Use log or square root of area if the relationship is nonlinear.

Robust standard errors (HAC/White’s correction)

Even if heteroscedasticity remains, use heteroscedasticity-robust standard errors (HC0, HC3).

Weighted Least Squares (WLS)

Assign weights inversely proportional to variance of residuals.

2. Handling Multicollinearity

Multicollinearity makes coefficient estimates unstable.

Detection:

Check Variance Inflation Factor (VIF) → VIF > 5 (or 10) signals collinearity.

Correlation heatmap between predictors.

Solutions:

Remove redundant predictors → e.g., drop area if it is highly correlated with number_of_rooms.

Combine correlated variables → e.g., create a single “size_index” instead of both area & rooms.

Regularization methods

Ridge Regression (L2) → shrinks coefficients, handles multicollinearity.

Lasso Regression (L1) → performs variable selection.

Elastic Net → balances both.

Principal Component Regression (PCR)

Use PCA to create orthogonal features before regression.

3. Ensuring a Robust Model

Cross-validation → use k-fold CV to ensure stability.

Check residual plots → after transformation, residuals should show no clear pattern.

Standardize/normalize features if using regularization.

Model comparison

Compare Linear Regression (with corrections) vs. Ridge/Lasso/Tree-based models.

Sometimes, non-linear models (Random Forest, Gradient Boosting) outperform regression in house price prediction.

Final Answer (Summary for Exam/Interview)

To address heteroscedasticity, I would log-transform the target variable (house price) and, if necessary, use robust standard errors or Weighted Least Squares. To address multicollinearity, I would first diagnose it using VIF, then either drop/recombine correlated features or switch to regularized regression methods like Ridge/Lasso. Finally, I’d validate the model with cross-validation to ensure stability and robustness.
